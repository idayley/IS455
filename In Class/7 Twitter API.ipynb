# Public APIs

```python
import requests, json

url = 'https://api.publicapis.org/entries'
res = requests.get(url)
json_data = json.loads(res.text)
json_clean = json.dumps(json_data, indent=2)
print(json_clean)
```

```python
import pandas as pd

#this is a quick way to get the data into a dataframe
# df = pd.DataFrame.from_dict(json_data['entries'])

#this is the manual way if the above way doesn't work
df = pd.DataFrame(columns=json_data['entries'][0].keys())
for entry in json_data['entries']:
    df.loc[entry['API']] = entry.values()
df.set_index('API', inplace=True)

df

```
